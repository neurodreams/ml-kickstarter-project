````python
import pandas as pd
````

````python
df = pd.read_csv("../data/Kickstarter_raw_data.csv")
````

````python
# Rate-sarakkeiden null-arvot

rate_cols = [col for col in df.columns if "rate" in col]

result_df = (
    pd.concat(
        {
            col: df.loc[df[col].isna(), "currency"].value_counts()
            for col in rate_cols
        },
        axis=1
    )
    .fillna(0)
    .astype(int)
)

result_df
````

````python
# Rivit, joista uupuu kaikki rate-tiedot

df.loc[
    df["static_usd_rate"].isna() &
    df["usd_exchange_rate"].isna() &
    df["fx_rate"].isna(),
    "currency"
].value_counts()
````

## Valuuttakurssit

* ```static_usd_rate```sisältää vähiten null-arvoja --> paras valinta(?)
* Yhdestäkään rivistä ei puutu *kaikkia* rate-sarakearvoja (static, fx, exchange)
